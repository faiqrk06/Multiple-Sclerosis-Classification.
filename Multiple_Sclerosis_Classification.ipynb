```python
#new technique
import numpy as np
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, LSTM, SimpleRNN, Embedding, Flatten
from tensorflow.keras.callbacks import EarlyStopping
from tensorflow.keras.preprocessing.sequence import pad_sequences

# Train a Random Forest classifier
rf_classifier = RandomForestClassifier()
rf_classifier.fit(X_train.reshape(X_train.shape[0], -1), y_train)

# Use the Random Forest predictions as additional features
rf_train_predictions = rf_classifier.predict(X_train.reshape(X_train.shape[0], -1))
rf_test_predictions = rf_classifier.predict(X_test.reshape(X_test.shape[0], -1))

# Combine the original features with Random Forest predictions
X_train_combined = np.column_stack((X_train.reshape(X_train.shape[0], -1), rf_train_predictions))
X_test_combined = np.column_stack((X_test.reshape(X_test.shape[0], -1), rf_test_predictions))

# Pad sequences for input data
X_train_padded = pad_sequences(X_train_combined, padding='post', maxlen=len(X_train_combined[0]))
X_test_padded = pad_sequences(X_test_combined, padding='post', maxlen=len(X_test_combined[0]))

# Reshape data for LSTM input
X_train_padded = X_train_padded.reshape((X_train_padded.shape[0], 1, X_train_padded.shape[1]))
X_test_padded = X_test_padded.reshape((X_test_padded.shape[0], 1, X_test_padded.shape[1]))

# Define a simple neural network with LSTM and RNN layers
model = Sequential()
model.add(LSTM(8, activation='relu', input_shape=(X_train_padded.shape[1], X_train_padded.shape[2])))
#model.add(Dropout(0.3))  # Add dropout with a suitable rate
model.add(Dense(64, activation='relu'))
model.add(Dense(32, activation='relu'))
#model.add(Dropout(0.2))  # Add dropout with a suitable rate
model.add(Dense(1, activation='sigmoid'))

# Compile the model
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

# Train the neural network using the combined features with early stopping
history = model.fit(X_train_padded, y_train, epochs=50, batch_size=32, validation_data=(X_test_padded, y_test))

# Evaluate the model on the test set
y_pred = model.predict(X_test_padded)
y_pred_binary = np.round(y_pred)
accuracy = accuracy_score(y_test, y_pred_binary)

# Calculate precision, recall, and F1-score
precision = precision_score(y_test, y_pred_binary)
recall = recall_score(y_test, y_pred_binary)
f1 = f1_score(y_test, y_pred_binary)

# Print the metrics
print(f"Accuracy: {accuracy}")
print(f"Precision: {precision}")
print(f"Recall: {recall}")
print(f"F1-score: {f1}")
```

```python
import numpy as np
import tensorflow as tf
from tensorflow.keras.models import Model, Sequential
from tensorflow.keras.layers import Input, Dense, Dropout
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.callbacks import ModelCheckpoint
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import confusion_matrix, accuracy_score, log_loss

# Assuming you have X_train_std, X_test_std, y_train, y_test from your previous code

# Standardize the data
scaler = StandardScaler()
X_train_std_scaled = scaler.fit_transform(X_train)
X_test_std_scaled = scaler.transform(X_test)

# ANN model
def create_ann_model(input_shape):
    ann_model = Sequential()
    ann_model.add(Dense(53, activation='tanh', kernel_initializer='normal', input_shape=(input_shape,),
                kernel_regularizer=tf.keras.regularizers.l2(l=0.01)))
    ann_model.add(Dropout(0.1))
    ann_model.add(Dense(53, activation='tanh', kernel_regularizer=tf.keras.regularizers.l2(l=0.01)))
    ann_model.add(Dropout(0.1))
    return ann_model

# Hybrid model
def create_hybrid_model(ann_model):
    ann_input = Input(shape=(X_train_std_scaled.shape[1],))
    ann_output = ann_model(ann_input)

    # Combine the output of ANN
    final_output = Dense(1, activation='sigmoid')(ann_output)

    hybrid_model = Model(inputs=ann_input, outputs=final_output)
    hybrid_model.compile(optimizer=Adam(lr=0.01), loss='binary_crossentropy', metrics=['accuracy'])

    return hybrid_model

# Create ANN model
ann_model = create_ann_model(input_shape=X_train_std_scaled.shape[1])

# Create hybrid model
hybrid_model = create_hybrid_model(ann_model)

# Save the best model during training based on validation accuracy
checkpoint = ModelCheckpoint("best_model.h5", monitor='val_accuracy', save_best_only=True, mode='max')

# Train hybrid model with history tracking
history = hybrid_model.fit(X_train_std_scaled, y_train,
                           batch_size=32, epochs=50,
                           validation_data=(X_test_std_scaled, y_test),
                           callbacks=[checkpoint])

# Load the best model
best_model = tf.keras.models.load_model("best_model.h5")

# Evaluate best model
best_model.evaluate(X_test_std_scaled, y_test)

# Make predictions
y_pred_best = best_model.predict(X_test_std_scaled)
y_pred_best = (y_pred_best > 0.5)

# Calculate metrics
#confmat_best = confusion_matrix(y_test, y_pred_best)
accuracy_best = accuracy_score(y_test, y_pred_best)
cross_entropy_best = log_loss(y_test, y_pred_best)

print("Best Model:")
#print("Confusion Matrix:\n", confmat_best)
print("Accuracy:", accuracy_best)
print("Log Loss:", cross_entropy_best)

# Print history
print("Training History:")

```
